```python
# ALL NECESSARY IMPORTS
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# Generating our Dataset
dataset = make_blobs(n_samples=200,
                    centers=4,
                    n_features=2,
                    cluster_std=1.6,
                    random_state=50)
```

```python
points = dataset[0]
```

```python
# importing kmeans
from sklearn.cluster import KMeans
```

```python
#Creating kmeans Objects
kmeans = KMeans(n_clusters=4)
```

```python
#Fit the kmeans object to the datset
kmeans.fit(points)
```

```python
plt.scatter(dataset[0][:,0], dataset[0][:,1])
```

```python
clusters = kmeans.cluster_centers_
```

```python
#Printing all clusters
print(clusters)
```

```python
x_km = kmeans.fit_predict(points)
```

```python

plt.scatter(points[x_km==0,0], points[x_km==0,1], s=50, color='green')
plt.scatter(points[x_km==1,0], points[x_km==1,1], s=50, color='red')
plt.scatter(points[x_km==2,0], points[x_km==2,1], s=50, color='blue')
plt.scatter(points[x_km==3,0], points[x_km==3,1], s=50, color='yellow')

#Centroids

plt.scatter(clusters[0][0], clusters[0][1], marker='*', s=200, color='black')
plt.scatter(clusters[1][0], clusters[1][1], marker='*', s=200, color='black')
plt.scatter(clusters[2][0], clusters[2][1], marker='*', s=200, color='black')
plt.scatter(clusters[3][0], clusters[3][1], marker='*', s=200, color='black')
```
